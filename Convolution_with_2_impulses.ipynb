```python
impulse = thinkdsp.read_wave('mika_clap_left_in_space_6.wav')
impulse.ys = impulse.ys[.85*44100:2.85002*44100]
impulse.make_audio()
```

```python
impulse2 = thinkdsp.read_wave('mika_clap_right_in_space_5.wav')
impulse2.ys = impulse2.ys[.85*44100:2.85002*44100]
impulse2.make_audio()
```

```python
m_wave = thinkdsp.read_wave('92002__jcveliz__violin-origional.wav')
# m_wave = m_wave.segment(start = 5)
m_wave.make_audio()
m_wave = m_wave | thinkdsp.Wave(numpy.zeros(44100),framerate = 44100)
m_wave.make_audio()
```

```python
m_wave.segment(start = 10, duration = 5)
m_wave.plot()
```

```python
imp_spec = impulse.make_spectrum()
imp_spec2 = impulse2.make_spectrum()
m_spec = m_wave.make_spectrum()
```

```python
def boxcar(duration): 
    boxcar = numpy.append(numpy.ones(44100*duration/2), numpy.zeros(44100*duration/2)+.000000001)
    tiny_box = numpy.append(numpy.ones(500)/500,numpy.zeros(500)+.000000001)
    filt1 = numpy.convolve(boxcar, tiny_box)
    filt2 = numpy.convolve(filt1, tiny_box)
    filt2 = filt2[:duration*44100]
    filt_wave = thinkdsp.Wave(filt2, 44100)
    return filt_wave
```

```python
imp_spec2.plot()
imp_spec2.hs = .5*imp_spec2.hs
imp_spec2.plot()
```

```python
# Tiny_waves = numpy.zeros(int(len(m_wave)/44100))
durati0n=len(impulse)/44100

yolo = int(len(m_wave)/44100)-1
some_wave = m_wave.segment(start = 0, duration=durati0n)
some_wave.ys = some_wave.ys * boxcar(durati0n).ys
some_spec = some_wave.make_spectrum()

balance = 0.0 #impulse2/impulse1
ratio_rate = .2
swag = (1.0-balance)*some_spec.hs * imp_spec.hs + (balance)*some_spec.hs*imp_spec2.hs

convolved_spec = thinkdsp.Spectrum(swag, framerate = 44100)
convolved_spec.plot()
convolved_wave = convolved_spec.make_wave()

convolved_wave = convolved_wave | thinkdsp.Wave(numpy.zeros(44100*4),framerate = 44100)
out_wave = convolved_wave

for i in range(1, yolo):
    some_wave = m_wave.segment(start = 0+i, duration=durati0n)
    print(some_wave.ys)
    some_wave.ys = some_wave.ys * boxcar(durati0n).ys
    some_spec = some_wave.make_spectrum()

    hs = (1-balance)*some_spec.hs * imp_spec.hs + (balance)*some_spec.hs*imp_spec2.hs
    convolved_spec = thinkdsp.Spectrum(hs, framerate = 44100)
    convolved_wave = convolved_spec.make_wave()
    
    convolved_wave = thinkdsp.Wave(numpy.zeros(44100*i),framerate = 44100)| convolved_wave | thinkdsp.Wave(numpy.zeros(44100*(yolo-i)),framerate = 44100)
    out_wave = out_wave + convolved_wave
    balance += ratio_rate

out_wave.normalize()
out_wave.plot()
some_wave.make_audio()
```
