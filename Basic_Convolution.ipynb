```python
import thinkdsp 
import thinkplot
import numpy 
import scipy
from scipy.io import wavfile
```

```python
#This is the wave recorded by microphones. Assuming the impulse response is something like a loud clap. 
m_wave = thinkdsp.read_wave('92002__jcveliz__violin-origional.wav')
m_wave.make_audio()
```

```python
#this is the recording of the impulse responses. Can be coppied for however long 
impulse = thinkdsp.read_wave('180961__kleeb__gunshots.wav')
impulse = impulse.segment(start=0.26, duration=500.0)
impulse.normalize()
impulse.plot()
impulse.make_audio()
# impulse_2 = wavfile.read('something else')
```

```python
m_wave.ys = m_wave.ys[:len(impulse)]
m_wave.normalize()
im_spec = impulse.make_spectrum()
m_spec = m_wave.make_spectrum()
len(im_spec.hs), len(m_spec.hs)
```

```python
out_wave = (im_spec*m_spec).make_wave()
```

```python
out_wave.normalize()
out_wave.make_audio()
```
